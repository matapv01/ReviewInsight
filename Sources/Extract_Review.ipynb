# 2. Sử dụng Selenium với Microsoft Edge

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from webdriver_manager.microsoft import EdgeChromiumDriverManager
import time

# Cấu hình Microsoft Edge
edge_options = Options()
edge_options.add_argument("--headless")  # Chạy không giao diện (tùy chọn)
edge_options.add_argument("--disable-gpu")

# Sử dụng Service để tạo WebDriver với WebDriverManager
service = Service(EdgeChromiumDriverManager().install())
driver = webdriver.Edge(service=service, options=edge_options)

# Truy cập trang sản phẩm Amazon
product_url = "https://www.amazon.com/Apple-iPhone-Pro-128GB-Graphite/dp/B08PNP5YGV/ref=sr_1_2?crid=OMB56TNKWF9G&dib=eyJ2IjoiMSJ9.LZY2BcPdz3eqKvYZ4E3g7fyrXl1W3WGl9Jghz7e7cUJVwFDwSBL_0C5-qjM40e3gg5snMMMSE0sHIg-ZbDJ8T4TCPpmj8ZloB7k4BXtQUcdWvY1FkEoXJPW2_fff3Q8Gvgbv95nuqaOE9efURW_oWlacDQl9qkNdM1Yf1M0Q9iuz6KqZHLD-MFikvvz8uxSRYE3s2jdxrZpYuPIPCaHw5aoaW9bNDGvh7QeEYWEV88w.MrFEpCW6nzg-oQQe_gmW5GI4ksi-NfBZa9F2iesboSY&dib_tag=se&keywords=iphone%2B12&qid=1726153787&sprefix=iphone%2B12%2Caps%2C302&sr=8-2&th=1"  # URL của sản phẩm (thay đổi thành sản phẩm bạn muốn)
driver.get(product_url)

# Đợi trang tải hoàn toàn
time.sleep(5)

# Cuộn trang để tải thêm đánh giá
last_height = driver.execute_script("return document.body.scrollHeight")
reviews = []

while True:
    # Lấy tất cả đánh giá của khách hàng
    review_elements = driver.find_elements("xpath", '//div[@data-hook="review"]//span[@data-hook="review-body"]//span')
    for review in review_elements:
        reviews.append(review.text)
    
    # Cuộn trang xuống
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Đợi trang tải thêm nội dung
    time.sleep(3)
    
    # Kiểm tra xem có thêm đánh giá mới không
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# In ra danh sách đánh giá
for i, review in enumerate(reviews, 1):
    print(f"Đánh giá {i}: {review}\n")

# Đóng trình duyệt
driver.quit()
```

# 3. Lưu đánh giá vào tệp CSV

```python
import csv

# Lưu đánh giá vào tệp CSV
with open("amazon_reviews.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["STT", "Đánh giá"])

    for i, review in enumerate(reviews, 1):
        writer.writerow([i, review])

print("Đánh giá đã được lưu vào amazon_reviews.csv")
```

# 4. Làm sạch file đánh giá và lưu vào file mới

```python
import csv

# Mở file CSV và tạo danh sách chứa các đánh giá hợp lệ
valid_reviews = []
with open("amazon_reviews.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    next(reader)  # Bỏ qua tiêu đề
    for row in reader:
        if row[1].strip() and "Read more" not in row[1]:  # Loại bỏ dòng trống và "Read more"
            valid_reviews.append(row)

# Lưu lại các đánh giá hợp lệ vào file CSV mới
with open("cleaned_amazon_reviews.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["STT", "Đánh giá"])  # Ghi tiêu đề
    writer.writerows(valid_reviews)

print(f"Đã lưu {len(valid_reviews)} đánh giá hợp lệ vào cleaned_amazon_reviews.csv")
```
