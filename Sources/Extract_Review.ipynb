# V√≠ d·ª• v·ªÅ selenium

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from webdriver_manager.microsoft import EdgeChromiumDriverManager

# C·∫•u h√¨nh Edge
edge_options = Options()
edge_options.add_argument("--headless")  # Ch·∫°y ·ªü ch·∫ø ƒë·ªô headless (t√πy ch·ªçn)
edge_options.add_argument("--disable-gpu")

# S·ª≠ d·ª•ng Service ƒë·ªÉ t·∫°o WebDriver v·ªõi WebDriverManager
service = Service(EdgeChromiumDriverManager().install())

# Kh·ªüi t·∫°o WebDriver cho Microsoft Edge v·ªõi service v√† options
driver = webdriver.Edge(service=service, options=edge_options)

# Truy c·∫≠p trang web
driver.get("https://www.dantri.com.vn")

# In ti√™u ƒë·ªÅ c·ªßa trang web
print(driver.title)

# ƒê√≥ng tr√¨nh duy·ªát
driver.quit()
```

# 2. S·ª≠ d·ª•ng Selenium v·ªõi Microsoft Edge

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from webdriver_manager.microsoft import EdgeChromiumDriverManager
import time

# C·∫•u h√¨nh Microsoft Edge
edge_options = Options()
edge_options.add_argument("--headless")  # Ch·∫°y kh√¥ng giao di·ªán (t√πy ch·ªçn)
edge_options.add_argument("--disable-gpu")

# S·ª≠ d·ª•ng Service ƒë·ªÉ t·∫°o WebDriver v·ªõi WebDriverManager
service = Service(EdgeChromiumDriverManager().install())
driver = webdriver.Edge(service=service, options=edge_options)

# Truy c·∫≠p trang s·∫£n ph·∫©m Amazon
product_url = "https://www.amazon.com/Apple-iPhone-Pro-128GB-Graphite/dp/B08PNP5YGV/ref=sr_1_2?crid=OMB56TNKWF9G&dib=eyJ2IjoiMSJ9.LZY2BcPdz3eqKvYZ4E3g7fyrXl1W3WGl9Jghz7e7cUJVwFDwSBL_0C5-qjM40e3gg5snMMMSE0sHIg-ZbDJ8T4TCPpmj8ZloB7k4BXtQUcdWvY1FkEoXJPW2_fff3Q8Gvgbv95nuqaOE9efURW_oWlacDQl9qkNdM1Yf1M0Q9iuz6KqZHLD-MFikvvz8uxSRYE3s2jdxrZpYuPIPCaHw5aoaW9bNDGvh7QeEYWEV88w.MrFEpCW6nzg-oQQe_gmW5GI4ksi-NfBZa9F2iesboSY&dib_tag=se&keywords=iphone%2B12&qid=1726153787&sprefix=iphone%2B12%2Caps%2C302&sr=8-2&th=1"  # URL c·ªßa s·∫£n ph·∫©m (thay ƒë·ªïi th√†nh s·∫£n ph·∫©m b·∫°n mu·ªën)
driver.get(product_url)

# ƒê·ª£i trang t·∫£i ho√†n to√†n
time.sleep(5)

# Cu·ªôn trang ƒë·ªÉ t·∫£i th√™m ƒë√°nh gi√°
last_height = driver.execute_script("return document.body.scrollHeight")
reviews = []

while True:
    # L·∫•y t·∫•t c·∫£ ƒë√°nh gi√° c·ªßa kh√°ch h√†ng
    review_elements = driver.find_elements("xpath", '//div[@data-hook="review"]//span[@data-hook="review-body"]//span')
    for review in review_elements:
        reviews.append(review.text)
    
    # Cu·ªôn trang xu·ªëng
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # ƒê·ª£i trang t·∫£i th√™m n·ªôi dung
    time.sleep(3)
    
    # Ki·ªÉm tra xem c√≥ th√™m ƒë√°nh gi√° m·ªõi kh√¥ng
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# In ra danh s√°ch ƒë√°nh gi√°
for i, review in enumerate(reviews, 1):
    print(f"ƒê√°nh gi√° {i}: {review}\n")

# ƒê√≥ng tr√¨nh duy·ªát
driver.quit()
```

# 3. L∆∞u ƒë√°nh gi√° v√†o t·ªáp CSV

```python
import csv

# L∆∞u ƒë√°nh gi√° v√†o t·ªáp CSV
with open("amazon_reviews.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["STT", "ƒê√°nh gi√°"])

    for i, review in enumerate(reviews, 1):
        writer.writerow([i, review])

print("ƒê√°nh gi√° ƒë√£ ƒë∆∞·ª£c l∆∞u v√†o amazon_reviews.csv")
```

# 4. L√†m s·∫°ch file ƒë√°nh gi√° v√† l∆∞u v√†o file m·ªõi

```python
import csv

# M·ªü file CSV v√† t·∫°o danh s√°ch ch·ª©a c√°c ƒë√°nh gi√° h·ª£p l·ªá
valid_reviews = []
with open("amazon_reviews.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    next(reader)  # B·ªè qua ti√™u ƒë·ªÅ
    for row in reader:
        if row[1].strip() and "Read more" not in row[1]:  # Lo·∫°i b·ªè d√≤ng tr·ªëng v√† "Read more"
            valid_reviews.append(row)

# L∆∞u l·∫°i c√°c ƒë√°nh gi√° h·ª£p l·ªá v√†o file CSV m·ªõi
with open("cleaned_amazon_reviews.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["STT", "ƒê√°nh gi√°"])  # Ghi ti√™u ƒë·ªÅ
    writer.writerows(valid_reviews)

print(f"ƒê√£ l∆∞u {len(valid_reviews)} ƒë√°nh gi√° h·ª£p l·ªá v√†o cleaned_amazon_reviews.csv")
```
